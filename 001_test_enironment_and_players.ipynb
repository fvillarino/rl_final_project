# Importar entorno y familiarizarse

```python
from boardgame2 import ReversiEnv
import numpy as np
```

### Instanciación y reset

```python
board_shape = 8
env = ReversiEnv(board_shape=board_shape)
(board, player) = env.reset()
print(board)
print(player)
```

### Jugadas válidas

```python
env.get_valid((board, player))
```

### Pasos

```python
(board, player), reward, done, _ = env.step([2,4])
print(board, player, done)
```

# Implementar una partida completa

Muestreando el esapcio de acción con env.action_space.sample

Notar que el estado es (board, player)

```python
done = False
(board, player) = env.reset()
while not done:
    action = env.action_space.sample()
    print(board)
    print(f'acion: {action}')
    
    print(f'player: {player}')
    (board, player), reward, done, _ = env.step(action)
    
```

# Implementar sample_valid_actions que muestrea uniformemente las acciones válidas

```python
def sample_valid_actions(state):
    # np.argwhere junto con env.get_valid y randint solucionan el problema en pocas lineas pero puede usar otra estrategia
    valid_actions = np.argwhere(env.get_valid(state) == 1)
    return valid_actions[np.random.randint(len(valid_actions))]
```
